```python
import matplotlib.pyplot as plt
import cv2
```

```python
img=cv2.imread("bmw-logo.png")
newImg=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
# Using matplot lib to display the image
plt.imshow(img)
plt.show()
plt.imshow(newImg)
plt.show()
```

```python
# Using OpenCV to display the image
img=cv2.imread("Dog.png")
```

```python
cv2.imshow("Dog Face",img)
cv2.waitKey(0) # waiting for unlimited time
cv2.destroyAllWindows()
```

```python
# Opening a grayscale image
img2=cv2.imread("Dog.png",cv2.IMREAD_GRAYSCALE)
cv2.imshow("Grayscale Dog",img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
