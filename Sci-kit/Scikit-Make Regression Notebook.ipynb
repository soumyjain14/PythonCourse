```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
```

```python
X,y=make_regression(n_samples=1000,n_features=2,n_targets=1,noise=10,random_state=1)
```

```python
plt.style.use("seaborn")
```

```python
plt.subplot(2,2,1)
plt.scatter(X[:,0],y,c='green')
plt.show()
plt.subplot(2,2,2)
plt.scatter(X[:,1],y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model=model.fit(X,y)
```

```python
model.predict([X[0]])
```
