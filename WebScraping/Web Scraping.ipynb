```python
from urllib.request import urlopen
```

```python
bmw_link='https://en.wikipedia.org/wiki/BMW'
```

```python
bmw_html=urlopen(bmw_link)
```

```python
bmw_get=bmw_html.read()
```

```python
bmw_get
```

```python
from bs4 import BeautifulSoup as soup
```

```python
bmw_soup=soup(bmw_get,'html.parser')
```

```python
bmw_soup
```

```python
bmw_soup.findAll("h1",{})
```

```python
tables=bmw_soup.findAll("table",{"class":"wikitable"})
```

```python
len(tables)
```

```python
bmw_table=tables[0]
```

```python
bmw_table
```

```python
headers=bmw_table.findAll("th",{})
```

```python
headers[0].text
```

```python
for i in headers:
    print(i.text, end ='')
```

```python
heading_list=[i.text[:-1] for i in headers]
```

```python
heading_list
```

```python
row_data=bmw_table.findAll("tr",{})[1:]
```

```python
first_row=row_data[0].findAll("td",{})
```

```python
first_row[0].text[:-1]
```

```python
for i in first_row:
    print(i.text[:-1])
```

```python
first_row=[i.text[:-1] for i in first_row]
```

```python
first_row
```

```python
heading_list
```

```python
tables=[]
for row in row_data:
    current_row=[]
    for data in row.findAll("td",{}):
        current_row.append(data.text[:-1])
    tables.append(current_row)
```

```python
filename='BMW.csv'
with open(filename,"w",encoding='UTF-8') as f:
    header_string=",".join(heading_list)
    header_string += '\n'
    f.write(header_string)
    
    for row in tables:
        row_string = ",".join(row)
        row_string += '\n'
        f.write(row_string)
    
```
