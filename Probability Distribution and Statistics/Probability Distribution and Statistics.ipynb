#### Multivariate Normal Distribution

```python
# Bi-variate
# We need2 things to define multivariate distribution covariance matrix& Mean matrix
Mean=np.array([0.0,0.0])
cov=np.array([[1,0],[0,1]])

Mean2=np.array([5.0,6.0])
cov2=np.array([[1.1,0.5],[0.5,0.8]])

dist=np.random.multivariate_normal(Mean,cov,1000) # Equal distribution for x and y
# sigma x =1 xy=yx=0 and y=1 so equal in all
dist2=np.random.multivariate_normal(Mean2,cov2,1000)
# sigma x=1.1 , xy=yx=0.5 and y=0.8, more in x, so greater spread in x direction
```

```python
dist.shape
```

```python
plt.scatter(dist[:,0],dist[:,1])
plt.scatter(dist2[:,0],dist2[:,1])
plt.show()
```
