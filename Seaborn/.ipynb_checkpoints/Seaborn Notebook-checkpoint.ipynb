```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
tips.head()
```

```python
plt.figure(figsize=(4,4))
plt.style.use("seaborn")
sns.barplot(x='sex' , y='total_bill' ,data=tips )
plt.title("Gender vs Total Bill")
plt.show()
```

```python
plt.style.use("dark_background")
plt.figure(figsize=(4,4))
sns.countplot(x='time',data=tips)
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.boxplot(x="day",y='total_bill',data=tips)
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.boxplot(x="day",y='total_bill',data=tips,hue='sex')
plt.show()
```

```python
themes=plt.style.available
print(themes)
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(3,3))
sns.violinplot(x="day",y='total_bill',data=tips,hue='smoker',split=True)
plt.show()
```

```python
plt.figure(figsize=(3,3))
sns.distplot(tips['total_bill'])
plt.show()
```

```python
plt.figure(figsize=(3,3))
sns.kdeplot(tips['total_bill'])
plt.show()
```

```python
plt.figure(figsize=(1,1))
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg',color='green',height=4)
plt.show()
```

```python
plt.figure(figsize=(1,1))
sns.jointplot(x='total_bill',y='size',data=tips,kind='reg',color='orange',height=4)
plt.show()
```

```python
# Pair Plot
# Plotting in relation to one-another
sns.pairplot(tips,height=2,hue="sex")
```

```python
# using Pivot and Flights Data
flights_data=sns.load_dataset('flights')
```

```python
flights_data.head()
```

```python
flight_pivot=flights_data.pivot(index='month',columns='year',values='passengers')
```

```python

plt.figure(figsize=(10,6))
sns.heatmap(flight_pivot,annot=True,fmt="d",cmap="YlGnBu")
```
