```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.arange(10)
y1=x**2
y2=2*x+3
```

```python
print(x)
print(y1)
print(y2)
```

```python
themes=plt.style.available
print(themes)
```

```python
plt.style.use("seaborn-dark")
```

```python
plt.figure(figsize=(3,3))
plt.plot(x,y1,"ro-",label="Apple")
plt.plot(x,y2,color="green",label="Kiwi")
plt.legend()
plt.xlabel("Time in Years")
plt.ylabel("Prices of Fruits")
plt.title("Fruits prices vs Time")
plt.show()
```

```python
# Bar Graph
metal1=np.array([10,20,30])
metal2=np.array([12,17,33])
x_coordinates=np.array([0,1,2])*2
```

```python
plt.figure(figsize=(6,4))
plt.bar(x_coordinates,metal1,width=0.5,color='green',tick_label=["Iron","Silver","Gold"],label="Current Year")
plt.bar(x_coordinates+0.5,metal2,color='pink',width=0.5,label="Next Year")
plt.xlabel("Metals")
plt.ylabel("Prices")
plt.ylim(0,40)
plt.title("Prices of Metals over two years")
plt.legend()
plt.show()
```

```python
# Using Pie Charts
plt.style.use('seaborn')
subjects="Mathematics","Physics","Chemistry","English","Physical Education"
weightage=[97,63,65,92,69]
plt.title("Distribution of Marks Subject Wise")

colors = ['yellow', 'lightgreen', 'green', 'orange', 'red', 'lightsalmon', 'darkred']
plt.pie(weightage,labels=subjects, autopct='%1.1f%%',explode=(0.1,0,0,0.2,0),startangle=60,colors=colors,shadow=True)
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.show()
```

```python
# Standard Normal Distribution
# Generating random normal Numbers
x_random=np.random.randn(100)
# xsn=(X(value)-mean)/sigma
#xsn==x.random here
```

```python
sigma=8 # Standard Deviation
Mean=70
X=np.round(x_random*sigma+Mean)
```

```python
print(X)
```

```python
plt.style.use("seaborn")
plt.hist(X,color='blue',rwidth=15,label='Maths Marks')
plt.xlabel('Subjects')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
import pandas as pd
```

```python
dataset=pd.read_csv('movies.csv')
```

```python
dataset
```

```python
film_titles=list(dataset['Film'])
```

```python
type(film_titles)
```

```python
freq_count=dict()
```

```python
for titles in film_titles:
    freq_length=len(titles)
    if freq_count.get(freq_length) is None:
        freq_count[freq_length]=1
        
    else:
        freq_count[freq_length] +=1
        
  
```

```python
freq_count
```

```python
X=np.array(list(freq_count.values()))
```

```python
Y=np.array(list(freq_count.keys()))
```

```python
print(X,Y)
```

```python
plt.figure(figsize=(5,5))
plt.scatter(X,Y,c="red",)
plt.xlabel("No. of Movies")
plt.ylabel("Count of Characters of movies names")
plt.title("Count of Characters VS No. of them having same Count")
plt.show()
```
